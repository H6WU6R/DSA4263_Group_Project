```python
# Add this as a new code cell after the existing cells

import os
import pandas as pd

# Base directory for the datasets
base_dir = "/Users/leizhenyu/Library/CloudStorage/OneDrive-NationalUniversityofSingapore/DSA4263/hackathon/Phishing Email Dataset"
# List all files in the base_dir
for filename in os.listdir(base_dir):
    print(filename)
```

```python
files = ['Nigerian_Fraud.csv',
         'Nazario.csv',
         'SpamAssasin.csv',
         'CEAS_08.csv']
for file in files:
    file_path = os.path.join(base_dir, file)
    df = pd.read_csv(file_path)
    print(f"Dataset: {file}")
    print(df.head())
    # print(df.columns)
    print("\n")
```

```python
columns = ['sender', 'receiver', 'date', 'subject', 'body', 'urls', 'label']

dfs = []
for file in files:
    file_path = os.path.join(base_dir, file)
    df = pd.read_csv(file_path)
    # Reindex columns, fill missing columns with NaN
    df = df.reindex(columns=columns)
    dfs.append(df)

merged_df = pd.concat(dfs, ignore_index=True)
merged_df.head()
```
